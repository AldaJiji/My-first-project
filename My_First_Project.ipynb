## Imports

```python
#Standard imports
import pandas as pd
import matplotlib.pyplot as plt
```

## 1. Data collection

We will be working with the `currencies.csv`.

This csv file is obtained from https://www.kaggle.com/datasets/anurag629/currencies and contains information on currencies in different countries.

```python
#we Load csv file into df

df=pd.read_csv('currencies.csv')
df
```

## 2. Data preparation

```python
df.info()
```

```python
#We remove all missing values

df=df.dropna()
df.info()
```

## 3.  Data analysis and visualization

3.1 Top 5 currencies most used
 
  A lot of countries use the same currency, we are going to count the countries with the same currency and search for the top 5 currencies used in the most countries.
Afterwards, We will create a piechart with the `plt.pie()`, to visualize the top 5 currencies most used.

```python
# Count number of countries for each currency

#We create a new column 'Countries_Count'
#We use a lambda function which counts the number of countries for each row in the 'Countries' column
df['Countries_Count'] = df['Countries'].apply(lambda x: len(str(x).split(',')) if pd.notnull(x) else 0) 

df
```

```python
#We look for the 5 most used currencies 

#We sorts the df by the Countries_Count column in descending order and selects the top 5 rows
#Then We chose the Name column from these top 5 rows and converts it to a list, so we could plot later
top5_currencies=df.nlargest(5, 'Countries_Count')['Name'].tolist()

#we create a list of the top 5 Countries_Count too in order to plot
top5_countriescount=df.nlargest(5, 'Countries_Count')['Countries_Count'].tolist()
```

```python
#We plot the piechart

fig=plt.figure(figsize=(10,7))
plt.pie(top5_countriescount,labels=top5_currencies,labeldistance=None)

plt.title('The 5 most used currencies')
plt.legend(title=None,loc='upper right',fontsize='medium');
```

# 3.2 Multi-country currencies VS single-country currencies

As we saw, several countries use the same currency and certain countries are the only ones using their currency. Thus, We will compare the multi-country currencies vs the single-country currencies and visualize it.

```python
#Let's find all the single-country currencies and all the multi_country currencies

single_country_currencies=df[df['Countries_Count']==1]
multi_country_currencies=df[df['Countries_Count']!=1]
```

```python
#We plot both on the same figure to compare
plt.figure(figsize=(12, 6))

#We plot for single-country currencies
plt.bar(single_country_currencies['Name'], single_country_currencies['Countries_Count'], color='blue', label='Single-Country currencies')

# Plot for multi-country currencies
plt.bar(multi_country_currencies['Name'], multi_country_currencies['Countries_Count'], color='green', label='Multi-Country currencies')


plt.xlabel('Currency Name')
plt.ylabel('Countries_Count')
plt.title('Comparison of Countries_Count')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

# 3.3 Symbols
Countries denote their currencies with symbols. While some countries use special characters, others use real letters to denote their currencies. So, in this part we will see all the countries using special characters and all the countries using real letter as symbol of their currency. At the same time, we will be able to see the countries which use the same symbol to denote their currrency even though the latter have different names.

```python
# Separate data based on Symbol type
df_special_char = df[df['Symbol'].apply(lambda x: not x.isalnum())]
df_letters = df[df['Symbol'].apply(lambda x: x.isalnum())]
```

```python
# Plotting both graphs on the same figure
plt.figure(figsize=(12, 10))

# Plot for special characters
plt.bar(df_special_char['Name'], df_special_char['Symbol'], color='red', label='Special Characters')

# Plot for letters
plt.bar(df_letters['Name'], df_letters['Symbol'], color='blue', label='Letters')

plt.xlabel('Currencies')
plt.ylabel('Symbols')
plt.title('Symbols used for currencies')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.grid(True, axis='y', linestyle='--', alpha=0.7)  # Add grid lines for better comparison
plt.show()
```
