## Imports

```python
#Standard imports
import pandas as pd
import matplotlib.pyplot as plt
```

## 1. Data collection

We will be working with the `currencies.csv`.

This csv file is obtained from https://www.kaggle.com/datasets/anurag629/currencies and contains information on currencies in different countries.

```python
#we Load csv file into df

df=pd.read_csv('currencies.csv')
df
```

## 2. Data preparation and analysis

```python
df.info()
```

```python
#We remove all missing values

df=df.dropna()
df.info()
```

```python
#Look for list of countries
# Select value at row label 0,14,19,21 and 24 and column 'Countries'

uk_pound=df.at[0, 'Countries']
us_dollar=df.at[14, 'Countries']
aust_dollar=df.at[19, 'Countries']
nzd_dollar= df.at[21, 'Countries']
euro=df.at[24, 'Countries']
```

```python
#since we obtain strings of one element we are going to split each string into lists using split(',')
uk_pound_countries=uk_pound.split(',')
us_dollar_countries=us_dollar.split(',')
aust_dollar_countries=aust_dollar.split(',')
nzd_dollar_countries=nzd_dollar.split(',')
euro_countries=euro.split(',')
```

## 3. Visualization

We are going to create a piechart with the `plt.pie()`, to visualize the top 5 currencies most used.

```python
#We create a dataset to visualize the top 5 currencies most used


#Data
Currencies=['Uk Pound','US Dollar','Aust Dollar','NZD Dollar','Euro']
countries=[len(uk_pound_countries),len(us_dollar_countries),len(aust_dollar_countries),len(nzd_dollar_countries),len(euro_countries)]


fig=plt.figure(figsize=(10,7))
plt.pie(countries,labels=Currencies,labeldistance=None)

plt.title('The 5 most used currencies')
plt.legend(title=None,loc='upper right',fontsize='medium');

```
